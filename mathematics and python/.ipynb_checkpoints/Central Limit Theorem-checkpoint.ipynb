```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi, norm
from scipy.special import gamma
```

# Распределение хи

```python
# хи-распределение с количеством степеней свободы k = 5
k = 5
chi_rv = chi(k)

# выборка объёма 1000 из хи-распределения
sample = chi_rv.rvs(1000)
```

```python
# гистограмма выборки
plt.hist(sample, label='sample histogram', density=True, bins=25)

# теоретическая плотность распределения
x = np.linspace(0,5,1000)
pdf = chi_rv.pdf(x)

plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('fraction of samples / f(x)')
plt.xlabel('$x$')
plt.show()
```

# Оценка распределения выборочного среднего

Сгенерируем 1000 выборок объёма 5, 10 и 50:

```python
sample5 = [chi_rv.rvs(5) for _ in range(1000)]
sample10 = [chi_rv.rvs(10) for _ in range(1000)]
sample50 = [chi_rv.rvs(50) for _ in range(1000)]
```

Построим гистограммы получившихся выборок:

```python
fig, axs = plt.subplots(1, 3, constrained_layout=True, figsize=(10, 5))

axs[0].hist([sample.mean() for sample in sample5], density=True, bins=15)
axs[0].set_title('Распределение\nвыборочного среднего\nпо выборкам размера 5')
axs[0].set_xlabel('$x$')
axs[0].set_ylabel('fraction of samples')

axs[1].hist([sample.mean() for sample in sample10],density=True, bins=15)
axs[1].set_title('Распределение\nвыборочного среднего\nпо выборкам размера 10')
axs[1].set_xlabel('$x$')
axs[1].set_ylabel('fraction of samples')

axs[2].hist([sample.mean() for sample in sample50],density=True, bins=15)
axs[2].set_title('Распределение\nвыборочного среднего\nпо выборкам размера 50')
axs[2].set_xlabel('$x$')
axs[2].set_ylabel('fraction of samples')

plt.show()
```

## Теоретическое распределение выборочных средних

Для хи-распределения 
$$ \mu = \sqrt{2}\frac {Г((k+1)/2)} {Г(k/2)}, $$
$$ \sigma^{2} = k - \mu^{2}, $$
где $ Г(z) $ - гамма-функция, $ k $ - количество степеней свободы (параметр распределения)

Рассчитаем теоретические значения среднего и дисперсии:

```python
# Теоретическое среднее
mu = np.sqrt(2) * gamma((k + 1) / 2) / gamma(k / 2)

# Теоретическая дисперсия
sigma2 = k - mu ** 2
```

Построим теоретические распределения средних вместе с гистограммами:

```python
# Вид распределения средних объёма 5
norm_rv5 = norm(mu, np.sqrt(sigma2 / 5))

# Вид распределения средних объёма 10
norm_rv10 = norm(mu, np.sqrt(sigma2 / 10))

# Вид распределения средних объёма 50
norm_rv50 = norm(mu, np.sqrt(sigma2 / 50))
```

```python
fig, axs = plt.subplots(1, 3, constrained_layout=True, figsize=(10, 5))
x = np.linspace(1,4,100)

axs[0].hist([sample.mean() for sample in sample5], bins=15,density=True, alpha=0.5)
axs[0].plot(x, norm_rv5.pdf(x))
axs[0].set_title('Распределение\nсреднего\nпо выборкам размера 5')
axs[0].set_xlabel('$x$')
axs[0].set_ylabel('fraction of samples / f(x)')

axs[1].hist([sample.mean() for sample in sample10], bins=15, density=True, alpha=0.5)
axs[1].plot(x, norm_rv10.pdf(x))
axs[1].set_title('Распределение\nсреднего\nпо выборкам размера 10')
axs[1].set_xlabel('$x$')
axs[1].set_ylabel('fraction of samples / f(x)')

axs[2].hist([sample.mean() for sample in sample50], bins=15, density=True, alpha=0.5)
axs[2].plot(x, norm_rv50.pdf(x))
axs[2].set_title('Распределение\nсреднего\nпо выборкам размера 50')
axs[2].set_xlabel('$x$')
axs[2].set_ylabel('fraction of samples / f(x)')

plt.show()
```

# Выводы

С увеличением объёма выборки распределение выборочных средних всё лучше приближается нормальным.
Также при увеличении объёма выборки уменьшается дисперсия распределения средних.
